# 1-25. Image Manipulation:
Open an image, adjust brightness/contrast, rotate/flip, crop a specific area, and save the modified image. Color Space Conversion: Convert an image between RGB, grayscale, and other color spaces
like HSV. Analyze how different spaces highlight different features.

```python
import cv2
import matplotlib.pyplot as plt

# Open an image
image_path = '/content/mount-makiling-philippines-v0-09w2g0k8qi7a1.jpg' #change this if you have problem and upload your own picture
image = cv2.imread(image_path)

# Adjust brightness and contrast
alpha = 1.5  # Contrast control (1.0 means no change)
beta = 25    # Brightness control (0 means no change)
adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Rotate/flip
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
flipped_image = cv2.flip(image, 1)  # 1 for horizontal flip, 0 for vertical flip

# Crop a specific area (e.g., top-left 100x100 pixels)
crop_x, crop_y, crop_width, crop_height = 0, 0, 100, 100
cropped_image = image[crop_y:crop_y + crop_height, crop_x:crop_x + crop_width]

# Save modified images
cv2.imwrite('adjusted_image.jpg', adjusted_image)
cv2.imwrite('rotated_image.jpg', rotated_image)
cv2.imwrite('flipped_image.jpg', flipped_image)
cv2.imwrite('cropped_image.jpg', cropped_image)

# Display original and modified images
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB)), plt.title('Adjusted')
plt.show()
```

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Display original and converted images
plt.subplot(131), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(132), plt.imshow(gray_image, cmap='gray'), plt.title('Grayscale')
plt.subplot(133), plt.imshow(hsv_image), plt.title('HSV')
plt.show()
```
